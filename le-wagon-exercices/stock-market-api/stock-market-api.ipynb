```python
import requests
```

```python
url = "https://cloud.iexapis.com/stable/stock/aapl/chart/3m?token=sk_8e51a562e50d47daa50571d3e41b5f66"
```

```python
api_data = requests.get(url).json()
```

```python
apple_stock_df = pd.DataFrame.from_dict(api_data)
```

```python
apple_stock_df['date'] = pd.to_datetime(apple_stock_df['date'], format="%Y-%m-%d")
```

```python
apple_stock_df = apple_stock_df.set_index('date')
```

```python
apple_stock_df['close'].plot()
```

```python
apple_stock_df[['open', 'close', 'high', 'low']].plot()
```
